Rating Data

```python
ratings = pd.read_csv('/content/Ratings.csv')
```

```python
ratings.head()
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

Merging the tables


```python
ratings_with_books = ratings.merge(books, on ='ISBN')
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns = {'rating': 'number of ratings'}, inplace = True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on = 'title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number of ratings'] >= 50 ]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id', 'title'],inplace = True)
```

```python
final_rating.shape
```

Create the pivot table for better understanding of the solution

```python
book_pivot = final_rating.pivot_table(columns = 'user_id',index = 'title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot = book_pivot.fillna(0)
```

```python
book_pivot.shape
```

```python
type(book_pivot)
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

Model Implementation

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm ='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[54, :].values.reshape(1,-1), n_neighbors=6)
```

```python
suggestions
```

```python
book_pivot.index[54]
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index == 'Animal Farm'
```

```python
np.where(book_pivot.index =='Animal Farm')[0][0]
```

Main Function

```python
def recommend_book(book_name):
  book_id = np.where(book_pivot.index == book_name)[0][0]
  distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1), n_neighbors=6)

  for i in suggestions[0]:
    print(book_pivot.index[i])
```

```python
recommend_book("Animal Farm")
```
