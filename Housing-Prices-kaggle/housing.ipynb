Importing the dependencies

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error
```

Importing the data

```python
data = pd.read_csv(r'Datasets/train.csv')
```

```python
data
```

Data Preprocessing

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['Id',"Alley","PoolQC","Fence","MiscFeature","MasVnrType","MoSold","YrSold"],inplace=True)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
#Managing missing values
data['LotFrontage'].fillna(data['LotFrontage'].mean(),inplace=True)
data['MasVnrArea'].fillna(data['MasVnrArea'].mean(),inplace=True)
data['BsmtQual'].fillna(data['BsmtQual'].mode()[0],inplace=True)
data['BsmtCond'].fillna(data['BsmtCond'].mode()[0],inplace=True)
data['BsmtExposure'].fillna(data['BsmtExposure'].mode()[0],inplace=True)
data['BsmtFinType1'].fillna(data['BsmtFinType1'].mode()[0],inplace=True)
data['BsmtFinType2'].fillna(data['BsmtFinType2'].mode()[0],inplace=True)
data['Electrical'].fillna(data['Electrical'].mode()[0],inplace=True)
data['FireplaceQu'].fillna(data['FireplaceQu'].mode()[0],inplace=True)
data['GarageType'].fillna(data['GarageType'].mode()[0],inplace=True)
data['GarageYrBlt'].fillna(data['GarageYrBlt'].mean(),inplace=True)
data['GarageFinish'].fillna(data['GarageFinish'].mode()[0],inplace=True)
data['GarageQual'].fillna(data['GarageQual'].mode()[0],inplace=True)
data['GarageCond'].fillna(data['GarageCond'].mode()[0],inplace=True)
```

Splitting the values in X and Y

```python
X = data.drop(columns=['SalePrice'])
Y = data['SalePrice']
```

```python
print(X.shape)
print(Y.shape)
```

Splitting the data in the train test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape, X_test.shape,X.shape)
```

Model Implementation - XGBoost

```python
xgb_reg = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
```

```python
xgb_reg.fit(X_train,Y_train)
```

```python
xgb_reg.predict(X_test)
```

```python
# Accuracy of the model
accuracy_score = xgb_reg.score(X_test,Y_test)
print(accuracy_score)
```

```python
# Precision of the model
threshold = 200000
# Convert the true values and predicted values into binary (1 for above the threshold, 0 for below)
y_true_binary = (Y_test > threshold).astype(int)
y_pred_binary = (xgb_reg.predict(X_test) > threshold).astype(int)

# Now calculate precision
precision = precision_score(y_true_binary, y_pred_binary)
```

```python
print(precision)
```

```python
# Mean Squared Error
MSE = mean_squared_error(Y_test,xgb_reg.predict(X_test))
```

```python
print(MSE)
```
